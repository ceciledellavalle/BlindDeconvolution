```python
K1, K2, x_im, x_b, x_o1 = DataGen()
```

```python
# test de la d√©rivation
Nx,Ny  = x_im.shape
M      = K1.shape[0]//2
# Padding K
K1_pad = np.pad(K1, ((Nx//2-M,Nx//2-M),(Ny//2-M,Ny//2-M)),'constant')
K2_pad = np.pad(K2, ((Nx//2-M,Nx//2-M),(Ny//2-M,Ny//2-M)),'constant')
# Derivation
d      = np.zeros((3,3))
d[1,1] = 1
d_pad  = np.pad(d, ((Nx//2-1,Nx//2-1),(Ny//2-1,Ny//2-1)), 'constant')
```

```python
import time
```

```python
start = time.time()
convolve2d(d_pad,K1_pad,mode='same')
print(time.time()-start)
```

```python
start = time.time()
fftconvolve(d_pad,K1_pad,mode='same')
print(time.time()-start)
```
